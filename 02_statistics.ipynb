### What is hypothesis testing and confidence interval for both categorical and numerical variables?

#### Hypothesis Testing
Used to assess two statements (hypothesis) about a population which are mutually exclusive. The best statements is chosen based on its statistical significance.

![hypothesis](https://www.section.io/engineering-education/hypothesis-testing-data-science/nullatlernative.png)

#### Confidence Interval
Range of reasonable values for our parameter  

> Best estimate $\pm$ Margin of Error  

**Best estimate** = unbiased point estimate  
**Margin of error** = "a few" estimated standard errors  
**Estimated standard error** = $\sqrt{\frac{\hat{p}(1-\hat{p})}{n}} $

> 95% confidence interval for p is; $$ \hat{p} \pm 1.96 \cdot \sqrt{\frac{\hat{p}(1-\hat{p})}{n}} $$  
95% confidence level refers to our confidence in the statistical procedure that was used to make this interval

###### Assumptions
1. Best estimate is based on simple random sample; a representative subset of the population that have an equal probability of being chosen.
2. Large enough sample size, n > 30
