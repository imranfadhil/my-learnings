# Awareness

## Awareness of the concept of probability and its applications

### What is probability and its applications?
Probability - Numerical description of the likelihood of a future event to occur.
$$ P(X) = \frac{event}{sample\ space\ (outcomes)} $$  

It has a wide application in various aspects of life;
1. weather forecasting
2. bank loan applications
3. business strategies
4. games
5. ...

#### Fundamentals of Probability Theory  
**Experiment** - trials  
**Event & Complement** - occurence vs non occurence  
**Random Variables** - Event. Discrete vs Continuous.  
**Probability Values** - P(X)  
**Sample space** - all possible outcomes of an experiment. A union of the events and complements.  
**Venn Diagram** - aka Set Diagram  
**Stochastic process** - simply random process  

**Permutation** - count elements where order is important.  
Without repition, $_{n}Pm_r = \frac{n!}{(n-r)!}$   
With repition, $_{n}Pm_r = n^r $

**Combination** - count elements where order is not important.   
Without repetition, $  _{n}Cm_r = \frac{n!}{(n-r)!r!} $  
With repetition, $  _{n}Cm_r = \frac{(r + n - 1)!}{(n-1)!r!} $  
where:  
n = numer of elements in the collection  
r = number of elements taken from the collection.  

Probability with regards to permutation and combination:

$$ P(X) = \frac{permutation\ or\ combination}{total\ number\ of\ combination\ or\ permutation} $$  



## Able to define and describe the applications of probability distributions

### What type of probability distribution are there?

### Discrete Distribution



---

#### Binomial  

PMF, $ f(x) = p^x(1-p)^{1-x},\  x=0,1 $  
Mean, $ \mu_x = p $  
Variance, $ \sigma^2 = p(1-p) $  


---

#### Negative Binomial 

---

#### Geometric

PMF, $ f(x) = p(1-p)^{x-1},\ x=1, 2, ... $   
Mean, $ \mu = E[X] = \frac{1}{p} $  
Variance, $ \sigma^2 = V[X] = \frac{(1-p)}{p^2} $

---

#### Poisson

PMF, $ f(x) = \frac{e^{-\lambda}\lambda^x}{x!},\ x=0, 1, 2, ..., \infty $  
Mean, $ E[X] = \lambda $  
Varianec, $ V[X] = \lambda $

---

### Continuous Distribution



---

#### Normal
PDF, $ f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^\frac{-(x-\mu)^2}{2 \sigma^2}  $

How to check Normality?
1. Visualization; histogram/ boxplot
2. Q-Q plot
3. KS Test/ Shapiro-Wilk Test

**Q-Q Plot** plots theoretical quantiles vs data quantiles and normal distribution would have the points lie on the diagonal.

  


**Shapiro-Wilk Test** calculates a W statisttic which tests whether a random sample comes from normal distribution. Small W value indicates departure from normality.

$$ W = \frac{(\sum^n_{i=1}ax_i)^2}{\sum^n_{i=1}(x_i - \bar{x})^2} $$

---

#### Log-Normal
PDF, $ f(x) = \frac{1}{x \sigma\sqrt{2\pi}}e^\frac{-(ln\ x-\mu)^2}{2 \sigma^2}  $  
Mean, $ E[X] = e^\frac{\mu+\sigma^2}{2} $  
Variance, $ V[X] = (e^{\sigma^2}-1)e^{2\mu+\sigma^2} $  

---

#### Chi-squared
PDF, $ f(x) = \frac{1}{2^\frac{k}{2}\Gamma(\frac{k}{2})} x^{\frac{k}{2}-1}e^{-\frac{x}{2}} $  
where k = degree of freedom.  

Mean, $ E[X] = k $  
Variance, $ V[X] = 2k $

---

#### Exponential
PDF, $ f(x) = ve^{-vx}  $  
Mean, $ E[X] = \frac{1}{v} $  
Variance, $ V[X] = \frac{1}{v^2} $ 

Models;
1. Time until failure
2. Separation between random events

### What is the difference between CDF, PDF and PMF?

**Probability Mass Function** (PMF) for discrete distribution and **Probability Density Function** (PDF) for continuous distribution.

$$ f(x) = P(X=x) $$  
where,  
1. $ f(x) \ge 0 $ for all x in $R_x$  
2. $ \sum_{x\in R_x} f(x) = 1 $  

For continuous random variables, the probability of a single event occuring is 0, $ P(X=c) = 0 $  

**Cumulative Density Function** (CDF)

$$ F(x) = P(X \le x) = \sum_{u \le x} f(u) $$  

where,  
1. $ F(-\infty) = 0 $ 
2. $ F(\infty) = 1 $  
3. if $ x \lt y $, then $ F(x) \le F(y) $

### Bayes Theorem
$$ P(A|B) = \frac{P(B|A)P(A)}{P(B)} $$

### Central Limit Theorem

**Standard Notations**

Population mean: $\mu$  
Population standard deviation: $\sigma$


Sample mean: $\bar{x}$  
Sample standard deviation: s

---
CLT states that the distribution of $\bar{x}$ is approximately normal, $\bar{x}$ ~ $(\mu, \frac{\sigma}{\sqrt{n}})$. This approximation gets better when the sample size increase.

The mean of a random sample from a population always has a normal distribution and the mean is equal to the population mean. Regardless of the nature of the population distribution;
- discrete or continuous
- symmetric or skewed
- unimodal or multimodal

### Discrete Data

To check the probability of encountering COAL when drilling a 5000 ft deep well.

```python
df = utils.auto_dtypes(df)
print('COAL unique values:\n', df.COAL.value_counts())
print()
print('Total COAL datapoints: ', df.COAL.count())
print()
p = df.COAL.value_counts()[1]/ df.COAL.value_counts()[0]
q = 1 - p
print('Probability of encountering COAL: ', round(p,3))

n = 5000 # Total depth of a well?
mean = n * p
std = np.sqrt(n*p*q)
pb = binom(n, p)
x = np.arange(mean - 2*std, mean + 2*std, dtype='int')

# PDF plot 
pmf = pb.pmf(x)

sns.set_style('whitegrid')
plt.vlines(x ,0, pmf, colors='k', linestyles='-', lw=5)
plt.ylabel('Probability')
plt.xlabel('Intervals')
plt.show()

# CDF plot
cdf = pb.cdf(x)

sns.set_style('whitegrid')
plt.vlines(x ,0, cdf, colors='k', linestyles='-', lw=5)
plt.ylabel('Probability')
plt.xlabel('Intervals')
plt.show()
```
